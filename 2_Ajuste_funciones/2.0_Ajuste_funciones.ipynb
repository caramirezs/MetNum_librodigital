## Interpolación y aproximación

El ajuste de funciones es una técnica importante en la resolución de problemas numéricos en ingeniería y ciencias aplicadas. La idea es encontrar una función matemática que se ajuste a un conjunto de datos discretos, lo que permite hacer predicciones y estimaciones de valores desconocidos en el intervalo de los datos conocidos. Los diferentes métodos de ajuste de funciones incluyen la interpolación, la aproximación polinómica y la aproximación por mínimos cuadrados, entre otros. Estos métodos son útiles porque permiten aproximar funciones complejas con funciones más simples, lo que puede facilitar el análisis y la interpretación de datos. Además, el ajuste de funciones es esencial en la elaboración de modelos matemáticos para la simulación y predicción de fenómenos físicos y en la toma de decisiones en la industria y la investigación científica.

Interpolación es un método utilizado para construir una función que pasa por un conjunto de puntos conocidos. La idea básica es encontrar una función continua que se ajuste a los datos discretos en un intervalo específico. Se pueden utilizar diferentes técnicas para la interpolación, como los **polinomios de Lagrange** o **Newton**, o los **splines cúbicos**. La interpolación se utiliza comúnmente en la construcción de tablas de valores, la predicción de valores desconocidos de una función, la elaboración de modelos matemáticos, la predicción de resultados de pruebas y experimentos, entre otras aplicaciones en la ingeniería y la ciencia.

Por otro lado, la aproximación de funciones es un método utilizado para encontrar una función simple que se ajusta a una función más compleja. A menudo, es difícil o incluso imposible obtener una función analítica exacta para ciertos problemas, y en tales casos, la aproximación de funciones puede ser una alternativa útil. Se pueden utilizar diferentes técnicas para la aproximación de funciones, como la **aproximación polinómica**, la **aproximación por mínimos cuadrados** y **la aproximación de series de Fourier**, entre otras. La aproximación de funciones se utiliza comúnmente en la construcción de modelos matemáticos, la predicción de resultados de pruebas y experimentos, la optimización de procesos y la elaboración de algoritmos para el análisis de datos.

En cuanto a las aplicaciones en la ingeniería, un ejemplo común de interpolación es la creación de mapas de contorno para campos de temperatura, presión o cualquier otra propiedad física en un sistema. En la industria aeronáutica, por ejemplo, la interpolación se utiliza para predecir la resistencia aerodinámica de un avión en diferentes condiciones de vuelo.

Por otro lado, la aproximación de funciones se utiliza en la ingeniería eléctrica para diseñar circuitos eléctricos que cumplan con ciertas especificaciones. Por ejemplo, en la construcción de filtros, se utilizan aproximaciones de funciones para diseñar circuitos que puedan filtrar frecuencias específicas.

En resumen, tanto la interpolación como la aproximación de funciones son técnicas importantes en la resolución de problemas numéricos en ingeniería y ciencias aplicadas. Estos métodos se utilizan para aproximar funciones complejas con funciones más simples, lo que facilita el análisis y la interpretación de datos, y son fundamentales en la elaboración de modelos matemáticos para la simulación y predicción de fenómenos físicos.

A continuación se presenta una gráfica que ilustra la diferencia entre la **interpolación** y **la aproximación** de funciones. Se muestra un ejemplo de una función que tiene un comportamiento no lineal y que se ajusta a un conjunto de datos discretos. En la figura de la izquierda, se utiliza la interpolación con **splines cúbicos** para ajustar los datos y obtener una curva suave que *pase por todos los puntos*. Por otro lado, en la figura de la derecha, se utiliza la **aproximación polinómica de grado 3** para ajustar los datos y obtener una curva que se ajusta al comportamiento general de la función, pero *no necesariamente pasa por cada punto*.

La interpolación es útil cuando se requiere obtener una curva suave que pase por todos los puntos y se busca una alta precisión en los valores intermedios. Por otro lado, la aproximación es útil cuando se busca una simplificación del comportamiento de la función y se acepta una menor precisión en los valores intermedios. En general, la elección de uno u otro método dependerá del propósito específico de la aplicación y de las limitaciones y requisitos de precisión del problema en cuestión.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
import seaborn as sns

# Generar datos de ejemplo
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([0, 1, 2, 1, -1, 0, 1])

# Interpolación con splines cúbicos
cs = CubicSpline(x, y)
x_interp = np.linspace(0, 6, 50)
y_interp = cs(x_interp)

# Aproximación
p = np.polyfit(x, y, 3)
y_approx = np.polyval(p, x_interp)

# Crear gráficos
sns.set_style('darkgrid') # Agrega el estilo "dark" de Seaborn
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Gráfico de interpolación
axs[0].plot(x_interp, y_interp, label='Interpolación')
axs[0].scatter(x, y, label='Datos', color='red')
axs[0].set_title('Interpolación', fontsize=15)
axs[0].legend()

# Gráfico de aproximación
axs[1].plot(x_interp, y_approx, label='Aproximación')
axs[1].scatter(x, y, label='Datos', color='red')
axs[1].set_title('Aproximación', fontsize=15)
axs[1].legend()

for ax in axs:
    # Obtiene los límites inferior y superior de x y y de los ejes del objeto de ejes.
    xmin, xmax = ax.get_xlim()
    ymin, ymax = ax.get_ylim()
    # label size de los ejes
    ax.tick_params(labelsize=13, axis='x')
    ax.tick_params(labelsize=13, axis='y')
    # Agrega flechas a los ejes x e y para indicar su dirección.
    ax.annotate("", xy=(xmax, 0), xytext=(xmin, 0), arrowprops=dict(color='gray', width=1.5, headwidth=8, headlength=10))
    ax.annotate("", xy=(0, ymax), xytext=(0, ymin), arrowprops=dict(color='gray', width=1.5, headwidth=8, headlength=10))

# Mostrar gráficos
plt.show()
```
